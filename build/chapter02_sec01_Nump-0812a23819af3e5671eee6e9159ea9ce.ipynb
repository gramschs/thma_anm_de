## Section 2.1 — Introduction to NumPy

NumPy (**Num**erical **Py**thon) is the fundamental library for numerical computing in Python.
It provides an efficient multi-dimensional array type (`ndarray`) and a large collection of
mathematical functions that operate on these arrays.

### Why NumPy instead of plain Python lists?
- NumPy arrays store **one fixed data type** — this allows vectorised C-level operations
  that are orders of magnitude faster than Python loops.
- NumPy provides built-in support for **linear algebra**, statistics, random numbers, and more.

## 2.1.1 Creating NumPy Arrays

## 2.1.5 Broadcasting

When two arrays have **compatible but different shapes**, NumPy **broadcasts** the smaller
array across the larger one without making copies.

Rules (simplified):
1. Arrays are compared shape dimension by dimension, **from the right**.
2. A dimension of size **1** is stretched to match the other.
3. If dimensions are neither equal nor 1, an error is raised.

Study the examples below carefully.

````python
vec_row    = np.array([1, 2, 3])           # shape (3,)
vec_column = np.array([[1], [2], [3]])     # shape (3,1)

print(f'vec_column * mat33 =\n {vec_column * mat33}')
print()
print(f'vec_row * mat33 =\n {vec_row * mat33}')
print()
print(f'vec_row * vec_column =\n {vec_row * vec_column}')

# vec_column * mat33 → each ROW of mat33 is multiplied by the corresponding element of vec_column
# vec_row    * mat33 → each COLUMN of mat33 is multiplied by the corresponding element of vec_row
# vec_row * vec_column → outer product: results in a (3,3) matrix
````
